```python
import numpy as np

# Згенерувати числовий діапазон
start = 5
size_of_vec = 100

# Генерація з використанням linspace
a = np.linspace(start, size_of_vec)
print(f"Згенерувала {len(a)} чисел в діапазоні {a[0]} до {a[-1]}")

# Вивести кожне друге число згенерованого діапазону за допомогою циклу
homework("Кожне друге число:")
for i in range(1, len(a), 2):  # крок 2
    print(a[i], end=" ")

print("\n\nМожна зменшити кількість чисел до 5:")
a = np.linspace(start, size_of_vec, 5)
print(a)
homework("Збільшую кількість чисел до 100 і виведу перші та останні 10:")
a = np.linspace(start, size_of_vec, 100)
print(f"Перші 10 чисел: {a[:10]}")
print(f"Останні 10 чисел: {a[-10:]}")

# Виключення останнього значення
print("""
Перше та останнє значення завжди входять у згенерований діапазон. 
Відповідно при цьому вираховується крок.
Це можна змінити, виключивши останнє значення з визначення діапазону.
""")
print(np.linspace(start, size_of_vec, 5, endpoint=False))
homework("Крок числового ряду, коли останнє значення виключено:")
step = (size_of_vec - start) / 5
print(f"Крок (вручну): {step}")

# Порівняння автоматичного визначення кроку
_, sp_1 = np.linspace(start, size_of_vec, 5, endpoint=True, retstep=True)
_, sp_2 = np.linspace(start, size_of_vec, 5, endpoint=False, retstep=True)
print(f"Для порівняння, визначу крок автоматично, у нашому випадку це {sp_1} (з endpoint=True) та {sp_2} (з endpoint=False)")

homework(f"Крок з endpoint=True: {sp_1}, крок з endpoint=False: {sp_2}. Різниця пояснюється тим, що з `endpoint=False` останнє число виключено з розрахунку, тому крок більший.")
```

```python
import numpy as np

# Приклади масивів різної розмірності
examples = {
    'Скаляр': np.array(1), 
    'Вектор Рядок': np.array([1, 2, 3]), 
    'Вектор Стовпець': np.array([[1], [2], [3]]), 
    'Матриця': np.array([[1, 2], [3, 4], [5, 6]]),
    'Багатовимірна Матриця': np.array([[[111, 112], [121, 122]],
                                        [[211, 212], [221, 222]],
                                        [[311, 312], [321, 322]]])
}

# Виведення інформації для кожного прикладу
for key in examples:
    print(f"Розглядаємо приклад з {key}. \nЗначення x буде: \n", examples[key])
    print(f"Тип {key} x: ", type(examples[key]))
    print(f"Розмірність/вимір {key} x:", np.ndim(examples[key]))
    print(f"Форма {key} x: ", examples[key].shape)
    print(f"Кількість елементів х: {np.size(examples[key])}")
    print("-=-" * 10)

# Пояснення результату
homework("""Функція np.ndim повертає кількість вимірів масиву (rank). 
Вектор-рядок та вектор-стовпець мають однакову розмірність 1D (один вимір), тому np.ndim повертає 1 для обох. 
Форма (shape) вказує кількість елементів у кожному вимірі: наприклад, (3,) для рядка та (3, 1) для стовпця. 
Це значення shape пояснює відмінність між вектором-рядком і вектором-стовпцем.""")

# Завдання: Задайте власні вектори
my_vector_row = np.array([10, 20, 30, 40])
my_vector_column = np.array([[10], [20], [30], [40]])

print("Мій вектор-рядок:\n", my_vector_row)
print("Розмірність:", np.ndim(my_vector_row))
print("Форма:", my_vector_row.shape)

print("Мій вектор-стовпець:\n", my_vector_column)
print("Розмірність:", np.ndim(my_vector_column))
print("Форма:", my_vector_column.shape)
```

```python
import numpy as np

# Початкові приклади
examples = {
    '5 рядків': np.array([[1], [2], [3], [4], [5]]), 
    '5 стовпців': np.array([[1, 2, 3, 4, 5]]),
    '5 елементів у вкладеному масиві': np.array([[[1, 2, 3, 4, 5]]])
}

# Виведення shape та вигляду
for a in examples:
    print(f"Для {a} маю shape: {examples[a].shape}\nта вигляд:\n{examples[a]}")
    print("-" * 30)

# Власні масиви
my_arrays = {
    '1D масив (рядок)': np.array([10, 20, 30]),
    '2D масив (матриця)': np.array([[10, 20], [30, 40], [50, 60]]),
    '3D масив (куб)': np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]]),
    '4D масив': np.array([[[[1, 2], [3, 4]]]])
}

# Виведення власних прикладів
for b in my_arrays:
    print(f"Для {b} маю shape: {my_arrays[b].shape}\nта вигляд:\n{my_arrays[b]}")
    print("-" * 30)

homework("Використовуючи метод `shape`, можна легко визначати виміри та структуру масивів. Це особливо корисно для аналізу багатовимірних даних у математиці, статистиці чи машинному навчанні.")
```

```python
import numpy as np

# Вихідний масив
A = np.array([[3.4, 8.7, 9.9], 
              [1.1, -7.8, -0.7],
              [4.1, 12.3, 4.8],
              [7.0, -3.9, 5.1]])

# Виведення початкової інформації
print(f"""Масив:
{A}
Виміри: {A.shape}
Принцип номерації ТІЛЬКИ:
Якщо задати одне значення, то отримаємо цілий рядок:
    Перший рядок: {A[0]}
    Останній рядок: {A[-1]}
Для вибору стовпця (символ ':' вказує на вибір усіх елементів):
    Перший стовпець: {A[:, 0]}
    Останній стовпець: {A[:, -1]}
Для вибору конкретного елемента треба два індекси:
    Перший елемент: {A[0, 0]}
    Останній елемент: {A[-1, -1]}
""")

# Завдання: виведення елементів, які належать діагоналі
homework("Ппрактикуюсь в доступі до багатовимірного масиву, виводити елементи, які належать діагоналі.")

# Діагональні елементи
main_diagonal = [A[i, i] for i in range(min(A.shape))]
print(f"Елементи головної діагоналі: {main_diagonal}")

# Виведення елементів побічної діагоналі
secondary_diagonal = [A[i, -i-1] for i in range(min(A.shape))]
print(f"Елементи побічної діагоналі: {secondary_diagonal}")

# Додаткові зрізи
print(f"Головна діагональ за допомогою np.diagonal(): {np.diagonal(A)}")
print(f"Побічна діагональ за допомогою np.fliplr(A).diagonal(): {np.fliplr(A).diagonal()}")
```

```python
import numpy as np

# Вихідний масив
A = np.array([
    [11, 12, 13, 14, 15],
    [21, 22, 23, 24, 25],
    [31, 32, 33, 34, 35],
    [41, 42, 43, 44, 45],
    [51, 52, 53, 54, 55]
])

# Виведення початкової інформації
print(f"""Масив:
{A}
Виміри: {A.shape}
Беремо декілька рядків: 
{A[0:2]}
Два перші стовпці: 
{A[:, 0:2]}
Два останні рядки:
{A[3:, :]}
Беремо підматрицю (нижні елементи): 
{A[-3:, -3:]}
""")

# Завдання: Витягуємо підматриці 2х2
homework("Практикуюсь у витягуванні елементів, беру матрицю 2х2 з нижнього та верхнього кутка матриці A.")

# Верхній лівий кут
top_left = A[:2, :2]
print(f"Підматриця 2х2 із верхнього лівого кута:\n{top_left}")

# Нижній правий кут
bottom_right = A[-2:, -2:]
print(f"Підматриця 2х2 із нижнього правого кута:\n{bottom_right}")
```

```python
import numpy as np

# Створюємо масив
A = np.arange(28).reshape(4, 7)

# Виводимо початкову інформацію
print(f"""Масив:
{A}
Виміри: {A.shape}
Беремо кожен другий елемент першого рядка: 
{A[0, ::2]}
Беремо кожен третій елемент першого рядка: 
{A[0, ::3]}
Кожен другий елемент у кожному другому рядку: 
{A[::2, ::2]}
Останні три елементи в кожному другому рядку:
{A[::2, 4:]}
Можна повністю змінити порядок елементів (дзеркально обернемо):
{A[::-1, ::-1]}
""")

# Завдання: Попрактикуйтесь у витягуванні елементів
homework("Практикуюсь у витягуванні елементів, будь-яка робоча комбінація.")

# Комбінації витягування
# Центральна частина матриці
center = A[1:3, 2:5]
print(f"Центральна частина матриці (2x3):\n{center}")

# Четвертий елемент кожного рядка
fourth_elements = A[:, 3]
print(f"Четвертий елемент кожного рядка:\n{fourth_elements}")

# Рядки в зворотному порядку
reversed_rows = A[::-1, :]
print(f"Рядки в зворотному порядку:\n{reversed_rows}")

# Перші три рядки та кожен другий елемент
subset = A[:3, ::2]
print(f"Перші три рядки, кожен другий елемент:\n{subset}")
```

```python
import numpy as np

# Створення масиву
A = np.arange(50)

# Умова для парних чисел
all_even = (A % 2 == 0)

# Комплексна умова для вибору чисел
complex_condition = (A < 30) & (A / 5 < 2)

# Виведення результатів
print(f"""Спочатку потрібно створити умову за якою здійснювати вибірку.
Результат умови буде або {all_even[0]} для {A[0]} або {all_even[1]} для {A[1]}.
Після підставляння умови пошуку парних чисел в масив отримаємо:
{A[all_even]}
Або застосувавти більш комплексну умову з двох порівнянь:
{A[complex_condition]}
""")

# Завдання: Вивести всі непарні числа з масиву
homework("Виведжу всі непарні числа з даного масиву.")

# Умова для непарних чисел (які не є парними)
all_odd = (A % 2 != 0)

# Виведення непарних чисел
print(f"Непарні числа в масиві: {A[all_odd]}")
```
